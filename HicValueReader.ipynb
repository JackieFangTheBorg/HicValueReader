```python
import glob
import numpy as np
import pandas as pd
```

```python
path='E:\\JLD\\Desktop\\py_workspace\\txtreader\\files'
```

```python
results=[]
for txt in glob.glob(path+'\\'+'\*.txt'):
    lines=open(txt,'r').readlines()
    flen=len(lines)-1
    index=0
    for i in range(flen):
        if '* Hic' in lines[i]:
            index=i
    Wordlist = lines[index].split(',')
    HicSection=Wordlist[0].strip('*')
    HicSection[5:]
    filename=txt.split('\\')
    filename=filename[-1]
    filename=filename.split('.')
    filename=filename[0]
    results.append([filename,HicSection[5:]])
results=np.array(results)
df = pd.DataFrame(results, columns=['file', 'Hic'])
```
